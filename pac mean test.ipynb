```python
import numpy as np
from algs_lib import *
```

```python
train_x, train_y, test_x, test_y, num_classes, train_len = gen_iris(normalize=True)
subsample_rate = int(0.5*train_len)

for ind in range(100):
    mean_with = []
    mean_without = []

    removed_train_x = np.delete(train_x, ind, 0)
    removed_train_y = np.delete(train_y, ind, 0)
    dist = []
    print(removed_train_x.shape)

    for i in range(100):
        shuffled_x, shuffled_y = shuffle(removed_train_x, removed_train_y)

        shuffled_x, shuffled_y = get_samples_safe(shuffled_x, shuffled_y, num_classes, subsample_rate)

        added_x = copy.deepcopy(shuffled_x)
        added_y = copy.deepcopy(shuffled_y)
        added_x[0] = train_x[ind]
        added_y[0] = train_y[ind]
    #     print(shuffled_x.shape, added_x.shape)

        output_orig = np.average(shuffled_x, axis=0)
        output_new = np.average(added_x, axis=0)

        mean_without.append(output_orig)
        mean_with.append(output_new)
        dist.append((output_orig - output_new)**2)
    true_mean_without = np.average(removed_train_x, axis=0)[0]
    var_without = np.var(removed_train_x, axis=0)
#     print(var_without)
#     assert(False)
    est_mean_without = np.average(mean_without, axis=0)[0]
    
    
#     print(f'diff in true mean without is {abs(true_mean_without - est_mean_without)}')
    true_mean_with = np.average(mean_without, axis=0)[0] * (subsample_rate-1)/subsample_rate + train_x[ind][0] * 1./subsample_rate
    est_mean_with = np.average(mean_with, axis=0)[0]
    var_with = np.var(mean_with, axis=0)
#     print(f'diff in true mean with is {abs(true_mean_with - est_mean_with)}')
    
    true_avg_dist = np.average(dist, axis=0)[0]
#     print(true_avg_dist)
#     print(true_mean_without, true_mean_with, var_without[0], var_with[0])
    upper_bound_dist = true_mean_without**2 + var_without[0] + true_mean_with**2 + var_with[0]
    print(true_avg_dist, upper_bound_dist)
```
